```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
clf = DecisionTreeClassifier(random_state=0)
iris = load_iris()
score = cross_val_score(clf, iris.data, iris.target, cv=10)
```

```python
print(score)
np.average(score)
```

```python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
#insert
df_data=pd.DataFrame(iris.data, columns=iris.feature_names)
df_data
```

```python
#insert
dtree_iteration=DecisionTreeClassifier(max_depth=None)#ปรับจน.Iteration
dtree_iteration.fit(df,y)#fit ค่าdata เข้ากับclassในtarget
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree_iteration, out_file=dot_data,
                feature_names=iris.feature_names,
                class_names=iris.target_names,
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
